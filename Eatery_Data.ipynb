```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep = '\t')
```

```python
#get all the Food items available in the Eatery
df.item_name.unique()
```

```python
#get the list of all the ordered item from the Eatery according to the quantity of order

item = df.groupby('item_name').sum()
item = item.sort_values(['quantity'], ascending=False)
item
```

```python
#get the highest ordered food item from eatery

item.head(1)
```

```python
#get the lowest ordered food item from eatery

item.tail(1)
```
